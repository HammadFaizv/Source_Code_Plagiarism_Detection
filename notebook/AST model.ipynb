## TRAINING THE MODEL

```python
final_df = pd.read_csv("final_ast_dataset.csv")
final_df.head()
```

```python
train, test = train_test_split(final_df, train_size=0.7)
```

```python
train.shape, test.shape
```

```python
train.drop(['filename_1_','filename_2_'], axis=1, inplace=True)
test.drop(['filename_1_','filename_2_'], axis=1, inplace=True)
```

```python
train.shape, test.shape
```

```python
rf_classifier = RandomForestClassifier(n_estimators=1500, max_depth=20, max_features = 'sqrt', n_jobs = -1, random_state=42)
```

```python
rf_classifier.fit(train.drop(['labels'], axis=1), train['labels'].values.ravel())
```

```python
pred_rf = rf_classifier.predict(test.drop(['labels'], axis=1))
```

```python
true_rf = test['labels'].values
```

```python
report = classification_report(true_rf, pred_rf)
```

```python
print(report)
```

```python
feature_importances = pd.DataFrame(rf_classifier.feature_importances_,
                                   index = train.drop(['labels'], axis=1).columns,
                                    columns=['importance']).sort_values('importance', ascending=False)

feature_importances
```

```python
import pickle
pkl_filename = "rf_ast_model.pkl" 
with open(pkl_filename, 'wb') as file:  
    pickle.dump(rf_classifier, file)
```
